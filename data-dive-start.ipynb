```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet('data/combined_files.parquet')
df.head()
```

```python
df.info(verbose = 1)
```

```python
plt.figure(figsize=(12,6))

plt.plot(df['knoxville_temperature_2m'])
plt.plot(df['memphis_temperature_2m'])
plt.plot(df['nashville_temperature_2m'])

plt.title('Average Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Average Temperature')
plt.legend(['Knoxville', 'Memphis', 'Nashville'])
plt.show()
```
